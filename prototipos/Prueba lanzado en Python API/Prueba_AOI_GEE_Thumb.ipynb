## Lanzar servidor
Al ejecutar, abre `http://127.0.0.1:8000` en el navegador.


```python
import subprocess, sys, time, webbrowser
proc = subprocess.Popen([sys.executable, 'gee_aoi_server.py'])
time.sleep(1)
url = 'http://127.0.0.1:8000'
print('Servidor lanzado:', url)
try:
    webbrowser.open(url)
except Exception:
    pass
print('Para parar el servidor: proc.terminate()')
```
